# Google Trends

```python
from pytrends.request import TrendReq
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pytrend = TrendReq()
```

```python
all_keywords = ['bitcoin', 'ethereum', 'litecoin', 'ripple', 'dogecoin']

keywords = ['Messi']

timeframes = ['today 5-y', 'today 12-m', 'today 3-m', 'today 1-m', 'now 7-d', 'now 1-d']
timeframe = timeframes[0]

cat = '0' #all categories. for example, for books, it would be 22
geo= '' #worldwide, 2 letter abbreviation for country
gprop = '' #web searches, image searches, youtube searches, news searches
```

```python
pytrend.build_payload(kw_list=keywords, timeframe=timeframe, cat=cat, geo=geo, gprop=gprop)
```

```python
df = pytrend.interest_over_time()
```

```python
df.head()
```

```python
df_region = pytrend.interest_by_region(resolution='COUNTRY', inc_low_vol=True, inc_geo_code=False)
df_region.head(10)
```

```python
df_region.reset_index().plot(x="geoName", y="Messi", figsize=(120, 10), kind="bar")
```
